```python
 # backtrader was imported, Cerebo engine was realized, the resulting cerebro instance was told to run a loop over the data
from __future__ import (absolute_import, division, print_function,
                        unicode_literals)
import backtrader as bt
if __name__ == '__main__':
    cerebro = bt.Cerebro()
    cerebro.broker.setcash(50000.0)

    print('Starting Portfolio Value: %.2f' % cerebro.broker.getvalue())

    cerebro.run()

    print('Final Portfolio Value: %.2f' % cerebro.broker.getvalue())
```

```python
# backtesting the individual stocks in our portfolio
class SmaCross(bt.SignalStrategy):
    def __init__(self):
        sma1, sma2 = bt.ind.SMA(period=10), bt.ind.SMA(period=20)
        crossover = bt.ind.CrossOver(sma1, sma2)
        self.signal_add(bt.SIGNAL_LONG, crossover)
        
# O - Realty income Corp
if __name__ == '__main__':
    cerebro = bt.Cerebro()
    cerebro.addstrategy(SmaCross)
    cerebro.broker.setcash(2000.0)
    cerebro.broker.setcommission(commission=0.001)

    O_data = bt.feeds.YahooFinanceCSVData(dataname='O.csv')
    
    cerebro.adddata(O_data)
    print('O Starting Portfolio Value: %.2f' % cerebro.broker.getvalue())
    cerebro.run()
    print('O Ending Portfolio Value: %.2f' % cerebro.broker.getvalue())
    cerebro.plot(iplot=False)
    
    
# KO - Coca-Cola 
if __name__ == '__main__':
    cerebro = bt.Cerebro()
    cerebro.addstrategy(SmaCross)
    cerebro.broker.setcash(2000.0)
    cerebro.broker.setcommission(commission=0.001)

    KO_data = bt.feeds.YahooFinanceCSVData(dataname='KO.csv')
    
    cerebro.adddata(KO_data)
    print('KO Starting Portfolio Value: %.2f' % cerebro.broker.getvalue())
    cerebro.run()
    print('KO Ending Portfolio Value: %.2f' % cerebro.broker.getvalue())
    cerebro.plot(iplot=False)
    
    
# PFE - Pfizer, Inc
if __name__ == '__main__':
    cerebro = bt.Cerebro()
    cerebro.addstrategy(SmaCross)
    cerebro.broker.setcash(2000.0)
    cerebro.broker.setcommission(commission=0.001)

    PFE_data = bt.feeds.YahooFinanceCSVData(dataname='PFE.csv')
    
    cerebro.adddata(PFE_data)
    print('PFE Starting Portfolio Value: %.2f' % cerebro.broker.getvalue())
    cerebro.run()
    print('PFE Ending Portfolio Value: %.2f' % cerebro.broker.getvalue())
    cerebro.plot(iplot=False)
    
    
# GILD - Gilead Sciences
if __name__ == '__main__':
    cerebro = bt.Cerebro()
    cerebro.addstrategy(SmaCross)
    cerebro.broker.setcash(2000.0)
    cerebro.broker.setcommission(commission=0.001)

    GILD_data = bt.feeds.YahooFinanceCSVData(dataname='GILD.csv')
    
    cerebro.adddata(GILD_data)
    print('GILD Starting Portfolio Value: %.2f' % cerebro.broker.getvalue())
    cerebro.run()
    print('GILD Ending Portfolio Value: %.2f' % cerebro.broker.getvalue())
    cerebro.plot(iplot=False)
    
    
# D - Dominion Energy    
if __name__ == '__main__':
    cerebro = bt.Cerebro()
    cerebro.addstrategy(SmaCross)
    cerebro.broker.setcash(2000.0)
    cerebro.broker.setcommission(commission=0.001)

    D_data = bt.feeds.YahooFinanceCSVData(dataname='D.csv')
    
    cerebro.adddata(D_data)
    print('D Starting Portfolio Value: %.2f' % cerebro.broker.getvalue())
    cerebro.run()
    print('D Ending Portfolio Value: %.2f' % cerebro.broker.getvalue())
    cerebro.plot(iplot=False)
    

# Main - Main Street Capital
if __name__ == '__main__':
    cerebro = bt.Cerebro()
    cerebro.addstrategy(SmaCross)
    cerebro.broker.setcash(2000.0)
    cerebro.broker.setcommission(commission=0.001)

    MAIN_data = bt.feeds.YahooFinanceCSVData(dataname='MAIN.csv')
    
    cerebro.adddata(D_data)
    print('MAIN Starting Portfolio Value: %.2f' % cerebro.broker.getvalue())
    cerebro.run()
    print('MAIN Ending Portfolio Value: %.2f' % cerebro.broker.getvalue())
    cerebro.plot(iplot=False)
    
    
# SBUX - Starbucks
if __name__ == '__main__':
    cerebro = bt.Cerebro()
    cerebro.addstrategy(SmaCross)
    cerebro.broker.setcash(2000.0)
    cerebro.broker.setcommission(commission=0.001)

    SBUX_data = bt.feeds.YahooFinanceCSVData(dataname='SBUX.csv')
    
    cerebro.adddata(SBUX_data)
    print('SBUX Starting Portfolio Value: %.2f' % cerebro.broker.getvalue())
    cerebro.run()
    print('SBUX Ending Portfolio Value: %.2f' % cerebro.broker.getvalue())
    cerebro.plot(iplot=False)

    
# TWTR - Twitter
if __name__ == '__main__':
    cerebro = bt.Cerebro()
    cerebro.addstrategy(SmaCross)
    cerebro.broker.setcash(2000.0)
    cerebro.broker.setcommission(commission=0.001)

    TWTR_data = bt.feeds.YahooFinanceCSVData(dataname='TWTR.csv')
    
    cerebro.adddata(TWTR_data)
    print('TWTR Starting Portfolio Value: %.2f' % cerebro.broker.getvalue())
    cerebro.run()
    print('TWTR Ending Portfolio Value: %.2f' % cerebro.broker.getvalue())
    cerebro.plot(iplot=False)
    

# WKHS - Workhorse
if __name__ == '__main__':
    cerebro = bt.Cerebro()
    cerebro.addstrategy(SmaCross)
    cerebro.broker.setcash(2000.0)
    cerebro.broker.setcommission(commission=0.001)

    WKHS_data = bt.feeds.YahooFinanceCSVData(dataname='WKHS.csv')
    
    cerebro.adddata(WKHS_data)
    print('WKHS Starting Portfolio Value: %.2f' % cerebro.broker.getvalue())
    cerebro.run()
    print('WKHS Ending Portfolio Value: %.2f' % cerebro.broker.getvalue())
    cerebro.plot(iplot=False)

    
# CTSH - Cognizant Technology Solutions 
if __name__ == '__main__':
    cerebro = bt.Cerebro()
    cerebro.addstrategy(SmaCross)
    cerebro.broker.setcash(2000.0)
    cerebro.broker.setcommission(commission=0.001)

    CTSH_data = bt.feeds.YahooFinanceCSVData(dataname='CTSH.csv')
    
    cerebro.adddata(CTSH_data)
    print('CTSH Starting Portfolio Value: %.2f' % cerebro.broker.getvalue())
    cerebro.run()
    print('CTSH Ending Portfolio Value: %.2f' % cerebro.broker.getvalue())
    cerebro.plot(iplot=False)
```
